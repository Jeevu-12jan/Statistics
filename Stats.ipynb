```python
#iii)
def weighted_mean(values, weights):
    if len(values) != len(weights):
        raise ValueError("The length of values and weights must be the same")
    
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    
    if total_weight == 0:
        raise ValueError("The sum of weights cannot be zero")
    
    return weighted_sum / total_weight
```

```python
#iv)

import math

def geometric_mean(numbers):
    if not numbers:  # Check if the list is empty
        return None
    product = 1
    n = len(numbers)
    
    for num in numbers:
        product *= num
    
    return product ** (1 / n)

# Example usage:
numbers = [4, 16, 64]
result = geometric_mean(numbers)
print("Geometric Mean:", result)
```

```python
#v)
def harmonic_mean(numbers):
    if not numbers:  # Check if the list is empty
        return None
    n = len(numbers)
    reciprocal_sum = sum(1 / num for num in numbers if num != 0)  # Sum of reciprocals, ignore zero
    if reciprocal_sum == 0:
        return None  # To avoid division by zero
    return n / reciprocal_sum

# Example usage:
numbers = [4, 16, 64]
result = harmonic_mean(numbers)
print("Harmonic Mean:", result)
```

```python
#vii)
def trimmed_mean(numbers, trim_percentage):
    if not numbers or trim_percentage < 0 or trim_percentage >= 50:
        return None  # Return None for empty list or invalid trim percentage

    # Sort the numbers
    sorted_numbers = sorted(numbers)

    # Calculate the number of elements to trim
    n = len(numbers)
    trim_count = int(n * trim_percentage / 100)

    # Remove the trimmed outliers
    trimmed_numbers = sorted_numbers[trim_count:n - trim_count]

    # Calculate and return the mean of the remaining numbers
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Example usage:
numbers = [1, 2, 3, 4, 100, 200, 300]
trim_percentage = 20  # Trim 20% from both ends
result = trimmed_mean(numbers, trim_percentage)
print("Trimmed Mean:", result)
```

```python
#iii)

import math

def calculate_variance_and_sd(numbers):
    if not numbers:  # Check if the list is empty
        return None, None  # Return None for both if list is empty
    
    n = len(numbers)
    mean = sum(numbers) / n  # Calculate the mean of the list
    
    # Calculate the variance using the formula
    variance = sum((x - mean) ** 2 for x in numbers) / n
    
    # Calculate the standard deviation (square root of variance)
    standard_deviation = math.sqrt(variance)
    
    return variance, standard_deviation

# Example usage:
numbers =int_list2
variance, standard_deviation = calculate_variance_and_sd(numbers)

print(f"Variance: {variance}")
print(f"Standard Deviation: {standard_deviation}")
```

```python
#v)

import math

def coefficient_of_variation(numbers):
    if not numbers:  # Check if the list is empty
        return None
    
    # Calculate the mean of the list
    mean = sum(numbers) / len(numbers)
    
    # Calculate the standard deviation
    variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
    standard_deviation = math.sqrt(variance)
    
    # Calculate the coefficient of variation
    cv = (standard_deviation / mean) * 100
    
    return cv

# Example usage:
numbers = int_list2
cv = coefficient_of_variation(numbers)
print("Coefficient of Variation (CV):", cv)
```

```python
#Q.3

class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        # Ensure that the lengths of outcomes and probabilities match
        if len(outcomes) != len(probabilities):
            raise ValueError("Outcomes and probabilities must have the same length.")
        
        # Ensure that the sum of probabilities is 1 (a valid probability distribution)
        if not (0 <= sum(probabilities) <= 1):
            raise ValueError("The sum of probabilities must be 1.")
        
        # Store outcomes and probabilities
        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """Calculate the expected value (mean) of the random variable."""
        return sum(x * p for x, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        """Calculate the variance of the random variable."""
        expected_value = self.expected_value()
        # Calculate E[X^2] first
        expected_value_squared = sum((x ** 2) * p for x, p in zip(self.outcomes, self.probabilities))
        return expected_value_squared - expected_value ** 2

# Example usage:
outcomes = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]

rv = DiscreteRandomVariable(outcomes, probabilities)

# Calculate expected value
expected_val = rv.expected_value()
print(f"Expected Value: {expected_val}")

# Calculate variance
variance_val = rv.variance()
print(f"Variance: {variance_val}")
```

```python
#Q.4

import random

class FairSixSidedDie:
    def __init__(self):
        # The outcomes of a fair six-sided die (1 to 6)
        self.outcomes = [1, 2, 3, 4, 5, 6]
        # All outcomes have equal probability (1/6)
        self.probabilities = [1/6] * 6
    
    def roll(self):
        """Simulate rolling the die and return a random outcome."""
        return random.choice(self.outcomes)
    
    def expected_value(self):
        """Calculate the expected value (mean) of the die roll outcomes."""
        return sum(x * p for x, p in zip(self.outcomes, self.probabilities))
    
    def variance(self):
        """Calculate the variance of the die roll outcomes."""
        expected_val = self.expected_value()
        expected_val_squared = sum((x ** 2) * p for x, p in zip(self.outcomes, self.probabilities))
        return expected_val_squared - expected_val ** 2

# Example usage:
die = FairSixSidedDie()

# Expected value (theoretical for a fair die)
expected_val = die.expected_value()
print(f"Expected Value (E[X]): {expected_val}")

# Variance (theoretical for a fair die)
variance_val = die.variance()
print(f"Variance (Var[X]): {variance_val}")
```

```python
#Q.5

import numpy as np

def generate_samples_and_stats(distribution, params, size=1000):
    """
    Generates random samples from the specified distribution and calculates their mean and variance.
    
    :param distribution: The name of the distribution ('binomial' or 'poisson').
    :param params: A tuple containing the distribution parameters (n, p) for binomial or (lambda_,) for Poisson.
    :param size: The number of samples to generate.
    :return: A tuple (mean, variance) of the generated samples.
    """
    if distribution == 'binomial':
        n, p = params  # Extract parameters for binomial distribution
        samples = np.random.binomial(n, p, size)
    elif distribution == 'poisson':
        lambda_ = params[0]  # Extract the parameter for Poisson distribution
        samples = np.random.poisson(lambda_, size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")
    
    # Calculate the mean and variance of the generated samples
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example usage:
# Binomial distribution with n=10 trials, p=0.5 probability of success
binomial_mean, binomial_variance = generate_samples_and_stats('binomial', (10, 0.5), size=10000)
print(f"Binomial Mean: {binomial_mean}, Binomial Variance: {binomial_variance}")

# Poisson distribution with lambda=5 (average rate)
poisson_mean, poisson_variance = generate_samples_and_stats('poisson', (5,), size=10000)
print(f"Poisson Mean: {poisson_mean}, Poisson Variance: {poisson_variance}")
```

```python
#Q.6)

import numpy as np

def generate_gaussian_samples(mean, std_dev, size):
    """
    Generates random samples from a Gaussian (normal) distribution.
    
    :param mean: The mean (μ) of the distribution.
    :param std_dev: The standard deviation (σ) of the distribution.
    :param size: The number of random samples to generate.
    :return: A tuple of (mean, variance, standard deviation) of the samples.
    """
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)
    
    # Calculate mean, variance, and standard deviation
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example usage:
mean = 0         # Mean (μ) of the distribution
std_dev = 1      # Standard deviation (σ) of the distribution
size = 10000     # Number of random samples to generate

# Generate the samples and calculate their statistics
mean_result, variance_result, std_dev_result = generate_gaussian_samples(mean, std_dev, size)

# Print the results
print(f"Generated Gaussian Distribution with Mean={mean}, Std Dev={std_dev}")
print(f"Sample Mean: {mean_result}")
print(f"Sample Variance: {variance_result}")
print(f"Sample Standard Deviation: {std_dev_result}")
```

```python
#Q.8)
import numpy as np
import scipy.stats as stats

def normal_pdf(x, mu, sigma):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution.
    
    :param x: The value at which to evaluate the PDF.
    :param mu: The mean (μ) of the normal distribution.
    :param sigma: The standard deviation (σ) of the normal distribution.
    :return: The value of the PDF at x.
    """
    # PDF of normal distribution formula
    pdf_value = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return pdf_value

# Example usage:
mu = 0          # Mean of the normal distribution
sigma = 1       # Standard deviation of the normal distribution
x = 0           # The point at which to evaluate the PDF

pdf_value = normal_pdf(x, mu, sigma)
print(f"PDF of normal distribution at x={x} (μ={mu}, σ={sigma}): {pdf_value}")
```

```python
#Q.11)
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data: Old layout and new layout (1 = purchase, 0 = no purchase)
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the number of successes (purchases) in both groups
successes = np.array([old_layout.sum(), new_layout.sum()])

# Sample sizes (number of visitors in each group)
sample_sizes = np.array([len(old_layout), len(new_layout)])

# Perform the z-test for proportions
z_stat, p_value = proportions_ztest(successes, sample_sizes, alternative='larger')

# Output the results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Conclusion based on p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new layout is significantly better.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the layouts.")
```

```python
#Q.15)
import numpy as np
from scipy import stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Calculate means for both layouts
    mean_a = np.mean(layout_a_clicks)
    mean_b = np.mean(layout_b_clicks)
    
    # Calculate standard deviations for both layouts
    std_a = np.std(layout_a_clicks, ddof=1)
    std_b = np.std(layout_b_clicks, ddof=1)
    
    # Calculate the sample sizes
    n_a = len(layout_a_clicks)
    n_b = len(layout_b_clicks)
    
    # Calculate the t-statistic
    t_stat = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))
    
    # Calculate the degrees of freedom using the formula for unequal variances (Welch's t-test)
    df = (((std_a**2 / n_a) + (std_b**2 / n_b))**2) / (((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1)))
    
    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df))
    
    # Output the results
    print(f"Mean of Layout A: {mean_a:.2f}")
    print(f"Mean of Layout B: {mean_b:.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"Degrees of freedom: {df:.3f}")
    print(f"P-value: {p_value:.3f}")
    
    # Conclusion based on the p-value
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the click-through rates of the two layouts.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the click-through rates of the two layouts.")

# Example data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
ab_test(layout_a_clicks, layout_b_clicks)
```

```python
#Q.16)
import numpy as np
from scipy import stats

def cholesterol_t_test(existing_drug_levels, new_drug_levels):
    # Calculate sample means for both groups
    mean_existing = np.mean(existing_drug_levels)
    mean_new = np.mean(new_drug_levels)
    
    # Calculate sample standard deviations for both groups
    std_existing = np.std(existing_drug_levels, ddof=1)
    std_new = np.std(new_drug_levels, ddof=1)
    
    # Calculate sample sizes
    n_existing = len(existing_drug_levels)
    n_new = len(new_drug_levels)
    
    # Calculate the t-statistic
    t_stat = (mean_existing - mean_new) / np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))
    
    # Calculate the degrees of freedom using Welch's t-test formula
    df = (((std_existing**2 / n_existing) + (std_new**2 / n_new))**2) / (((std_existing**2 / n_existing)**2 / (n_existing - 1)) + ((std_new**2 / n_new)**2 / (n_new - 1)))
    
    # Calculate the p-value (one-tailed test since we are testing if new drug is better)
    p_value = stats.t.cdf(t_stat, df)
    
    # Output the results
    print(f"Mean of Existing Drug: {mean_existing:.2f}")
    print(f"Mean of New Drug: {mean_new:.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"Degrees of freedom: {df:.3f}")
    print(f"P-value: {p_value:.3f}")
    
    # Conclusion based on the p-value
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis: The new drug is significantly more effective than the existing drug in reducing cholesterol levels.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the new and existing drugs in reducing cholesterol levels.")

# Example data: Cholesterol levels for existing drug and new drug
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform the t-test
cholesterol_t_test(existing_drug_levels, new_drug_levels)
```

```python
#Q.17)

import numpy as np
from scipy import stats

def intervention_t_test(pre_scores, post_scores):
    # Calculate the differences between post and pre intervention scores
    differences = np.array(post_scores) - np.array(pre_scores)
    
    # Calculate mean and standard deviation of the differences
    mean_difference = np.mean(differences)
    std_difference = np.std(differences, ddof=1)
    
    # Sample size
    n = len(differences)
    
    # Calculate the t-statistic
    t_stat = mean_difference / (std_difference / np.sqrt(n))
    
    # Degrees of freedom
    df = n - 1
    
    # Calculate the p-value (two-tailed test because we are testing for any difference)
    p_value = stats.t.sf(np.abs(t_stat), df) * 2  # Two-tailed test
    
    # Output the results
    print(f"Mean difference: {mean_difference:.2f}")
    print(f"Standard deviation of differences: {std_difference:.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"Degrees of freedom: {df}")
    print(f"P-value: {p_value:.3f}")
    
    # Conclusion based on p-value
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis: The intervention had a significant impact on the test scores.")
    else:
        print("Fail to reject the null hypothesis: The intervention did not have a significant impact on the test scores.")

# Example data: Pre- and post-intervention test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform the t-test
intervention_t_test(pre_intervention_scores, post_intervention_scores)
```

```python
#Q.21)

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate random data for age groups and voter preferences
np.random.seed(0)

# Define the possible values for age groups and voter preferences
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame to organize the data
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Display the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("\nFail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")
```
